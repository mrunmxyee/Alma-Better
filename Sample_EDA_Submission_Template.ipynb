## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

```python
# Chart - 1 visualization code
f, ax = plt.subplots(figsize=(10, 7))
plt.title('Number Of Terrorist Activities per Year')
sns.despine(f)
sns.distplot(data['Year'], bins=20,color="g")
```

The number of terrorist activities have gone up sharply after 2005.

```python
# Chart - 2 visualization code
plt.subplots(figsize=(15,6))
sns.countplot('AttackType',data=data,palette='autumn',order=data['AttackType'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Attacking Methods by Terrorists')
plt.show()
```

We came to know that the most frequently used attacking method by the terrorists is either bombing or explosion, followed by Arrested Assault and Assassination.Two major attack types are Bombing/Explosion and Armed Assaults which indicates the misuse of science and technology against Humanity.

```python
# Chart - 3 visualization code
plt.subplots(figsize=(15,6))
sns.countplot(data['Target_type'],palette='autumn',order=data['Target_type'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Favorite Targets')
plt.show()
```

Here is a graph of Favourite targets of the terrorists and as we can see, Private Citizens and Property tops the list. Then comes Business, Government, Police and Military which are close with very little margins. 

```python
# Chart - 4 visualization code
plt.subplots(figsize=(15,6))
sns.countplot('Region',data=data,palette='autumn',order=data['Region'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number Of Terrorist Activities By Region')
plt.show()
```

Terrorist acts in the Middle East and northern Africa have been seen to have fatal consequences. The Middle East and North Africa are seen to be the places of serious terrorist attacks. In addition, even though there is a perception that Muslims are supporters of terrorism, Muslims are the people who are most damaged by terrorist attacks. If you look at the graphics, it appears that Iraq, Afghanistan and Pakistan are the most damaged countries.

```python
# Chart - 5 visualization code
plt.subplots(figsize=(18,6))
sns.barplot(data['Country'].value_counts()[:15].index,data['Country'].value_counts()[:15].values,palette='autumn')
plt.title('Top Affected Countries')
plt.show()
```

Iraq is the top most affected country by terrorism. Pakistan and India come 2nd and 3rd. Giving close competition to India is Columbia at the 4th position. India , Pakistan and Afganistan has seen thousands of terrorists act which is a worrying factor.

```python
# Chart - 6 visualization code
pd.crosstab(data.Year, data.Region).plot(kind='area',figsize=(20,6))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
# Chart - 7 visualization code
topgroups=data[data['Group'].isin(data['Group'].value_counts()[1:11].index)]
pd.crosstab(topgroups.Year,topgroups.Group).plot(color=sns.color_palette('rainbow',10))
fig=plt.gcf()
fig.set_size_inches(18,6)
plt.show()
```

Taliban has created the most chaos in the recent years followed by FMLN. The decade 1980-1990 has seen some major activities in terms of terrorism as far as SL, IRA, FMLN is concerned. It was peacefull in the next decade as compared to the previous one.

```python
# Chart - 8 visualization code
plt.subplots(figsize=(15,6))
cas = data.groupby('Year').casualities.sum().to_frame().reset_index()
cas.columns = ['Year','casualities']
sns.barplot(x=cas.Year, y=cas.casualities, palette='Set3')
plt.xticks(rotation=90)
plt.title('Number Of Casualities Per Year')
plt.show()
```

The number of casualities are gradually increasing due to terrorism which is a serious state of concern.

```python
# Chart - 9 visualization code
pd.crosstab(data.Region,data.AttackType).plot.barh(stacked=True,width=1,color=sns.color_palette('RdYlGn',9))
fig=plt.gcf()
fig.set_size_inches(12,8)
plt.show()
```

```python
# Chart - 10 visualization code
coun_terror=data['Country'].value_counts()[:15].to_frame()
coun_terror.columns=['Attacks']
coun_kill=data.groupby('Country')['Killed'].sum().to_frame()
coun_terror.merge(coun_kill,left_index=True,right_index=True,how='left').plot.bar(width=0.9)
fig=plt.gcf()
fig.set_size_inches(18,6)
plt.show()
```

Terrorist acts in the Middle East and northern Africa have been seen to have fatal consequences. The Middle East and North Africa are seen to be the places of serious terrorist attacks. In addition, even though there is a perception that Muslims are supporters of terrorism, Muslims are the people who are most damaged by terrorist attacks. If you look at the graphics, it appears that Iraq, Afghanistan and Pakistan are the most damaged countries.

```python
# Chart - 11 visualization code
sns.barplot(data['Group'].value_counts()[1:15].values,data['Group'].value_counts()[1:15].index,palette=('Set3'))
plt.xticks(rotation=90)
fig=plt.gcf()
fig.set_size_inches(10,8)
plt.title('Terrorist Groups with Highest Terror Attacks')
plt.show()
```

```python
killData = data.loc[:,'Killed']
attackData = data.loc[:,'AttackType']
typeKillData = pd.concat([attackData, killData], axis=1)
typeKillFormatData = typeKillData.pivot_table(columns='AttackType', values='Killed', aggfunc='sum')
```

```python
# Chart - 12 visualization code
labels = typeKillFormatData.columns.tolist() # convert line to list
transpoze = typeKillFormatData.T # transpoze
values = transpoze.values.tolist()
fig, ax = plt.subplots(figsize=(15, 15), subplot_kw=dict(aspect="equal"))
plt.pie(values, startangle=90, autopct='%.2f%%')
plt.title('Types of terrorist attacks that cause deaths')
plt.legend(labels, loc='upper right', bbox_to_anchor = (1.3, 0.9), fontsize=15) # location legend
plt.show()
```

This is a pie chart indicating the various types of terrorist attacks that actually caused death. Our terrorism pie Chart is little messy but we can easily figure out the top terrorist attacks, where 44.58 percent bookings are from Armed Assault, 35.25 percent from Bombing/Explosion, 8.08 percent from Assassination and 6.20 percent from unknown sources.
